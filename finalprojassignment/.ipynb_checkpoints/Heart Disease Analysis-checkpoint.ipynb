```python
# start of project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart_disease.csv')
hd = df[df['HeartDisease'] == 1]
normal = df[df['HeartDisease'] == 0]
df
```

```python
# now find avgs for age, m/f, resting bp, chl, etc

df_hd = df[df['HeartDisease'] == 1]
df_normal = df[df['HeartDisease'] == 0]

print('Heart Disease:')
avg_age= df_hd['Age'].mean()
most_common_sex = df_hd['Sex'].mode()[0]
avg_rbp= df_hd['RestingBP'].mean()
avg_chl= df_hd['Cholesterol'].mean()

print("Avg age is",round(avg_age))
print("most common sex is",(most_common_sex))
print("Avg rsting bp is",round(avg_rbp))
print("Avg chlo is",round(avg_chl))

avg_age_norm= df_normal['Age'].mean()
most_common_sex_norm = df_normal['Sex'].mode()[0]
avg_rbp_norm= df_normal['RestingBP'].mean()
avg_chl_norm= df_normal['Cholesterol'].mean()

print('No HD:')
print("Avg age is",round(avg_age_norm))
print("most common sex is",(most_common_sex_norm))
print("Avg rsting bp is",round(avg_rbp_norm))
print("Avg chlo is",round(avg_chl_norm))

# why is chol so high for non hd?????????
```

```python
# firgure out how to bool or make func for hd vs no hd
                
a=df['Age']
b=df['RestingBP']
c=df['Cholesterol']
d=df['MaxHR']
e=df['Oldpeak']
f=df['FastingBS']
g=df['Sex']


fig, axes = plt.subplots(3, 2, figsize=(14, 10))
# axes[0,0].scatter(a,b, color='gray', label="age vs resting bp")
# axes[0,0].scatter(a,f, color='black',label='true model')
axes[0,0].scatter(a[df['HeartDisease']==0], b[df['HeartDisease']==0], color='green', label='No Heart Disease', alpha=0.3, s=16)
axes[0,0].scatter(a[df['HeartDisease']==1], b[df['HeartDisease']==1], color='red', label='Heart Disease',alpha=0.3, s=16)
axes[0,0].set_title('Age vs RBP')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('RBP')


# axes[0,1].scatter(a,c, color='yellow',label='true model')
# axes[0,1].scatter(a,c, color='yellow',label='true model')
axes[0,1].scatter(a[df['HeartDisease']==0], c[df['HeartDisease']==0], color='green', label='No Heart Disease',alpha=0.3, s=16)
axes[0,1].scatter(a[df['HeartDisease']==1], c[df['HeartDisease']==1], color='red', label='Heart Disease',alpha=0.3, s=16)
axes[0,1].set_title('Age vs CHOL')
axes[0,1].set_xlabel('Age')
axes[0,1].set_ylabel('CHOL')


# axes[1,0].scatter(a,d, color='red',label='true model')
# axes[1,0].scatter(a,d, color='red',label='true model')
axes[1,0].scatter(a[df['HeartDisease']==0], d[df['HeartDisease']==0], color='green', label='No Heart Disease',alpha=0.3, s=16)
axes[1,0].scatter(a[df['HeartDisease']==1], d[df['HeartDisease']==1], color='red', label='Heart Disease',alpha=0.3, s=16)
axes[1,0].set_title('Age vs Max HR')
axes[1,0].set_xlabel('Age')
axes[1,0].set_ylabel('Max HR')

axes[1,1].scatter(a[df['HeartDisease']==0], e[df['HeartDisease']==0], color='green', label='No Heart Disease',alpha=0.3, s=16)
axes[1,1].scatter(a[df['HeartDisease']==1], e[df['HeartDisease']==1], color='red', label='No Heart Disease',alpha=0.3, s=16)
axes[1,1].set_title('Age vs OldPeak')
axes[1,1].set_xlabel('Age')
axes[1,1].set_ylabel('OldPeak')


# axes[2,0].scatter(a,f, color='black',label='true model')
# axes[2,0].scatter(a,f, color='black',label='true model')
axes[2,0].scatter(a[df['HeartDisease']==0], f[df['HeartDisease']==0], color='green', label='No Heart Disease',alpha=0.3, s=16)
axes[2,0].scatter(a[df['HeartDisease']==1], f[df['HeartDisease']==1], color='red', label='Heart Disease',alpha=0.3, s=16)
axes[2,0].set_title('Age vs RBP')
axes[2,0].set_xlabel('Age')
axes[2,0].set_ylabel('RBP')

# axes[2,1].hist(g[df['HeartDisease']==0], label='No Heart Disease')
# axes[2,1].hist(g[df['HeartDisease']==1], label='Heart Disease')
# axes[2,1].set_title('Sex vs HD')
# axes[2,1].set_xlabel('Sex')
# axes[2,1].set_ylabel('Count')
# axes[2,1].legend()
import seaborn as sns

sns.countplot(x='Sex', hue='HeartDisease', data=df, ax=axes[2,1])

axes[2,1].set_title('Heart Disease by Sex')
axes[2,1].set_xlabel('Sex')
axes[2,1].set_ylabel('Count')
axes[2,1].legend(title='Heart Disease', labels=['No', 'Yes'])              
plt.tight_layout()
plt.show()

```

```python
from scipy import stats

plt.figure(figsize=(8, 5))

df_hd = df[df['HeartDisease'] == 1]
df_normal = df[df['HeartDisease'] == 0]

plt.scatter(df_hd['Age'],df_hd['MaxHR'], color = 'red',label='HD')
plt.scatter(df_normal['Age'],df_normal['MaxHR'], color = 'green',label=' Normal')
slope, intercept, r, p, _ = stats.linregress(df['Age'], df['MaxHR'])
plt.plot(df['Age'], slope * df['Age'] + intercept, color='black', label=f'y = {slope:.2f}x + {intercept:.2f}, r = {r:.2f}')

plt.xlabel('Age')
plt.ylabel('MAXHR')
plt.title('Age vs MHR')
plt.legend()
plt.tight_layout()
plt.show()


```

```python

men   = df[df['Sex'] == 'M']['MaxHR'].values   
women = df[df['Sex'] == 'F']['MaxHR'].values   

fig, ax = plt.subplots(figsize=(10, 6))

counts_m, edges_m, _ = ax.hist(
    men,
    bins=30,           
    alpha=0.55,        
    color='blue',
    label=f'Male (n={len(men)})',
    edgecolor='white'
)

counts_f, edges_f, _ = ax.hist(
    women,
    bins=30,
    alpha=0.55,
    color='pink',
    label=f'Female (n={len(women)})',
    edgecolor='white'
)

ax.set_xlabel('Max Heart Rate', fontsize=13)
ax.set_ylabel('Count', fontsize=13)
ax.set_title('Max HR by Sex', fontsize=14, fontweight='bold')
ax.legend()           
ax.grid(True, alpha=0.3)  

plt.tight_layout()  
plt.show()
```
