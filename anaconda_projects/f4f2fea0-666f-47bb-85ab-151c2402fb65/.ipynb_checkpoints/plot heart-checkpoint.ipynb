```python
# now find avgs for age, m/f, resting bp, chl, etc
df_hd = df[df['HeartDisease'] == 1]
df_normal = df[df['HeartDisease'] == 0]


avg_age= df['Age'].mean()
most_common_sex = df['Sex'].mode()[0]
avg_rbp= df['RestingBP'].mean()
avg_chl= df['Cholesterol'].mean()
print("Avg age is",round(avg_age))
print("most common sex is",(most_common_sex))
print("Avg rsting bp is",round(avg_rbp))
print("Avg chlo is",round(avg_chl))
```

```python
from scipy import stats

plt.figure(figsize=(8, 5))

df_hd = df[df['HeartDisease'] == 1]
df_normal = df[df['HeartDisease'] == 0]

plt.scatter(df_hd['Age'],df_hd['MaxHR'], color = 'red',label='HD')
plt.scatter(df_normal['Age'],df_normal['MaxHR'], color = 'green',label=' Normal')
slope, intercept, r, p, _ = stats.linregress(df['Age'], df['MaxHR'])
plt.plot(df['Age'], slope * df['Age'] + intercept, color='black', label=f'r = {r:.2f}')

plt.xlabel('Age')
plt.ylabel('MAXHR')
plt.title('Age vs MHR')
plt.legend()
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('heart_disease.csv')

df_hd     = df[df['HeartDisease'] == 1]
df_normal = df[df['HeartDisease'] == 0]

plt.figure(figsize=(8, 5))
plt.boxplot([df_normal['Oldpeak'], df_hd['Oldpeak']],
            tick_labels=['No Heart Disease', 'Heart Disease'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'))

plt.ylabel('Oldpeak')
plt.title('Oldpeak vs Heart Disease')
plt.tight_layout()
plt.show()
```
