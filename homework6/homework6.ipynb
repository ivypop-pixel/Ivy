```python
grid = np.zeros((8, 8), dtype=int)
grid[0::2, 0::2] = 1
print(grid)
```

```python
grid = np.zeros((8, 8), dtype=int) 
grid[0::2, 0::2] = 1
print(grid)
```

```python
def checkerboard():
    grid = np.zeros((8, 8), dtype=int)
    grid[0::2, 0::2] = 1
    grid[1::2, 1::2] = 1 
    return grid

print(checkerboard())
```

```python
def reverse_checkerboard():
    grid = np.zeros((8, 8), dtype=int)
    grid[0::2, 1::2] = 1
    grid[1::2, 0::2] = 1
    return grid
print(reverse_checkerboard())
```

```python
def expansion(arr, n):
    spac = ' ' * n
    return np.array([spac.join(s) for s in arr])
universe = np.array(['galaxy', 'clusters'])
print(expansion(universe,1))
print(expansion(universe, 2))
```
