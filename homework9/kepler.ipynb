```python
df = pd.read_csv('solar_system.csv',index_col=0)
orbital_period = df.iloc[10].astype(float)
perihelion     = df.iloc[8].astype(float)
aphelion       = df.iloc[9].astype(float)
print(orbital_period)
```

```python
# define perihelion, aphelion and semi-major axis in your own words.
# 1. plantest cloest point to sun, aphelion is farthest, semi is avg betweent those points
```

```python
semi_major_axis = (perihelion + aphelion) / 2
semi_major_axis
```

```python
new_row=df.loc["semi major axis"]=semi_major_axis
df = pd.concat([df, new_row])
print(df)
```

```python
a = df.iloc[-1].astype(float)
T = df.iloc[10].astype(float)

plt.scatter(a, T, label="Planets")
plt.xlabel("Semi-Major Axis (10^6 km)")
plt.ylabel("Orbital Period (days)")
plt.title("Semi-Major Axis vs Orbital Period")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def keplers_third_law(a, m):
    return m * a ** 1.5
popt, pcov = curve_fit(keplers_third_law, a, T, p0=[1])
m_fit = popt[0]
print(f"Fit: T = {m_fit:.2f} * a^1.5")
```
