```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave (x, A, w):
    """
    sine wave"""
    y= A * np.sin(w * x)
    return y
```

```python
# 0 to 2π with 1000 points
x = np.linspace(0, 2 *np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25,5))
for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w=1)
    axes[0].plot(x, y, label=f'A={A}')
axes[0].set_title("First subplot", fontsize=18)
axes[0].set_xlabel("x")
axes[0].set_ylabel("y = A*sin(x)")
axes[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axes[0].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'], fontsize=12)
axes[0].legend(title="Vary (w)", title_fontsize=12, fontsize=11)

# Subplot 2
for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A=1, w=w)
    axes[1].plot(x, y, label=f'ω = {w}')

axes[1].set_title("Second subplot", fontsize=18)
axes[1].set_xlabel("x", fontsize=14)
axes[1].set_ylabel("y = sin(ωx)", fontsize=14)
axes[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axes[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'], fontsize=12)
axes[1].legend(title="Vary (A)", title_fontsize=12, fontsize=11)

# 3 
for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f'A = {A},  ω = {w}')

axes[2].set_title("Third subplot", fontsize=18)
axes[2].set_xlabel("x", fontsize=14)
axes[2].set_ylabel("y = A · sin(ωx)", fontsize=14)
axes[2].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axes[2].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'], fontsize=12)
axes[2].legend(title="Vary the amplitude and frequency together", title_fontsize=12, fontsize=11)
plt.show()
```
