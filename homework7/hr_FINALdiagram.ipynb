```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/Users/ivanashashi/Desktop/Ivy_pythin_decal_sp26/course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
print("shape of data:", data.shape)
print("rows:", data.shape[0])
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
min_mag = np.min(magnitude)
print ("min magnitude is", min_mag)
max_mag = np.max(magnitude)
print ("max mag is", max_mag)
```

```python
min_mag_star = np.argmin(magnitude)
max_mag_star = np.argmax(magnitude)
print ("star is for min magn:", min_mag_star)
print ("star is for max magn:", max_mag_star)
```

```python
min_temp = np.min(temperature)
print ("min temp is", min_temp)
max_temp = np.max(temperature)
print ("max temp is", max_temp)
```

```python
min_temp_star = np.argmin(temperature)
max_temp_star = np.argmax(temperature)
print ("temp corresponding to star min:", min_temp_star)
print ("temp corresponding to star max:", max_temp_star)
```

```python
# 2.24 is brightest, 12.49 is dimmest
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
fig, axes = plt.subplots(1,1, figsize=(8, 6))
axes.scatter(log_temperature, log_magnitude)
brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)

x = log_magnitude 
y = log_temperature
axes.scatter(y[brightest], x[brightest],
             color="black",  marker='*', s=300, label="Brightest")

axes.scatter(y[dimmest],   x[dimmest],
             color="purple", marker="*", s=300, label="Dimmest")

axes.scatter(y[hottest],   x[hottest],
             color="green",  marker="*", s=300, label="Hottest")

axes.scatter(y[coldest],   x[coldest],
             color="orange", marker="*", s=300, label="Coldest")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

axes.legend(fontsize=11, title = "Info about Stars", title_fontsize=12)
axes.set_title('Hertzsprung-Russell Diagram', fontsize=16)
axes.set_xlabel("Temperature (log K)", fontsize=11)
axes.set_ylabel("Magnitude", fontsize=15)
plt.show()
```
