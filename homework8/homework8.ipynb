```python
# median latitude and median longitude

lat_mean = df["latitude"].median()
lon_mean = df["longitude"].median()
print("the median latitude is", lat_mean)
print("the median longitude is", lon_mean)
```

```python
# df['abs value distance to the median latitude']=df[abs(df['latitude']-lat_mean)] this was wrong bc i put the abs inside
df['abs_distance_to_median_lat']=abs(df['latitude']-lat_mean)
df.head(5)

```

```python
# absolute value distance to the median longitude
df['absval_distance_to_med_longitude']=abs(df['longitude']-lat_mean)
df.head(5)
```

```python
df.shape #now i have 6 columns
```

```python
# or the following print statements, please round your answer to two decimal places:
# • Print the state closest to the median latitude and by how much.
closest_lat = df.loc[df['abs_distance_to_median_lat'].idxmin(),'state']
closest_lat_val = df['abs_distance_to_median_lat'].min()
print(f"latitude close to the median is {closest_lat} by {closest_lat_val:.2f}.")


# • Print the state farthest from the median latitude and by how much.
fart_lat = df.loc[df['abs_distance_to_median_lat'].idxmax(),'state']
fart_lat_val = df['abs_distance_to_median_lat'].max()
print(f"latitude close to the median is {fart_lat} by {fart_lat_val:.2f}.")


# • Print the state closest to the median longitude and by how much.
closest_long = df.loc[df['absval_distance_to_med_longitude'].idxmin(), 'state']
closes_long_val=df['absval_distance_to_med_longitude'].min()
print(f"long close to the median is {closest_long} by {closes_long_val:.2f}.")

# • Print the state farthest from the median longitude and by how much.
fart_long = df.loc[df['absval_distance_to_med_longitude'].idxmin(), 'state']
fart_long_val=df['absval_distance_to_med_longitude'].min()
print(f"long close to the median is {fart_long} by {fart_long_val:.2f}.")

# Hint: You can round to two decimal places by attaching :.2f after your variables in f-strings.
# Example: print(f"{variable:.2f}").nn

```

```python
import numpy as np
# Create a figure with four subplots in a 2x2 grid.
df = sns.load_dataset("car_crashes")
fig, axes =plt.subplots(2, 2, figsize=(10, 8))

x= df["alcohol"]
y1= df["speeding"]
axes[0, 0].scatter(x, y1, color='red', label='Speeding')
axes[0, 0].set_xlabel('alcohol')
axes[0, 0].set_ylabel('speeding')
axes[0, 0].set_title('total vs speeding', color='black')
axes[0, 0].legend() 


z=df["not_distracted"]
y2=df["no_previous"]
axes[0,1].scatter(z,y2, color= 'blue')
axes[0,1].set_title('not distraced vs no_previous', color= 'black')
axes[0,1].set_xlabel('not_distracted')
axes[0,1].set_ylabel('no_previous)')
                      
y3=df['total']
m=df['ins_premium']
sns.histplot(m,ax=axes[1, 0], color='pink')
axes[1, 0].set_title('total crashes')
sns.boxplot(y3, ax=axes[1, 1], color='green')
axes[1, 1].set_ylabel('premium')
axes[1, 1].set_title('ins_premium')

plt.tight_layout() 
plt.show()

```
