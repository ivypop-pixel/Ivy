```python
df = pd.read_csv('solar_system.csv')
df = df.set_index("Attribute").T
df.index.name = "Planet"
df.reset_index(inplace=True)
df.columns.name = None
au_value = const.au.to(u.km).value
perihelion = df['Perihelion (10^6 km)'].astype(float).values * 1e6 / au_value * u.AU
aphelion   = df['Aphelion (10^6 km)'].astype(float).values   * 1e6 / au_value * u.AU
semi_major_axis = (perihelion + aphelion) / 2

aphelion_idx = df.columns.get_loc('Aphelion (10^6 km)')
df.insert(aphelion_idx + 1, 'semi_major_axis (AU)', semi_major_axis.value)

df.columns
print("Column names after insert:")
print(df.columns.tolist())
for planet, val in zip(df['Planet'], semi_major_axis):
    print(f"  {planet}: {val:.6f}")
```

```python
# Convert orbital period from days to years with astropy
orbital_period_day = df['Orbital Period (days)'].astype(float).values * u.day
orbital_period_yr  = orbital_period_day.to(u.year)
df['Orbital Period (days)'] = orbital_period_yr.value
df.rename(columns={'Orbital Period (days)': 'Orbital Period (years)'}, inplace=True)
print("Column names after rename:")
print(df.columns.tolist(), "\n")

print("Orbital Period (years) column:")
print(df['Orbital Period (years)'], "\n")

earth_period = df.loc[df['Planet'] == 'Earth', 'Orbital Period (years)'].values[0]
print(f"Earth's orbital period: {earth_period:.4f} years")
```

```python
# Rename the columns (e.g., "Diameter (km)" becomes "Diameter (AU)")
from astropy import units as u
from astropy import constants as const

au_value = const.au.to(u.km).value
new_columns = {
    'Diameter (km)' : 'Diameter (AU)',
    'Distance from Sun (10^6 km)':'Distance from Sun (AU)',
    'Perihelion (10^6 km)': 'Perihelion(AU)',
    'Aphelion (10^6 km)':'Aphelion (AU)',
    'semi_major_axis (km)':'semi_major_axis (AU)',    
}
scaled_cols = {
    'Distance from Sun (10^6 km)',
    'Perihelion (10^6 km)',
    'Aphelion (10^6 km)',
}

for old_name in new_columns:
    if old_name not in df.columns:
        continue
    values_km = df[old_name].astype(float)
    if old_name in scaled_cols:
        values_km = values_km * 1e6          # 10^6 km → km
    df[old_name] = (values_km.values * u.km).to(u.AU).value  
df.rename(columns=new_columns, inplace=True)
print(df.columns.tolist())
# Print a sample value from each converted column for your favorite planet.
planet = 'Earth'
row = df.loc[df['Planet'] == planet]
for new_name in new_columns.values():
    if new_name in df.columns:
        val = row[new_name].values[0]
        print(f"{planet} — {new_name}: {val:.6f} AU")
```
